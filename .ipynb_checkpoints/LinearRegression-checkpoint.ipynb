Scikit Learn accepts Numpy arrays as well as Pandas data frames

We split the data into train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30)
```

We instantiate the linear regression

```python
from sklearn.linear_model import LinearRegression
LinReg = LinearRegression()
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
columns = ['CRIM','ZN', 'INDUS' ,'CHAS' ,'NOX', 'RM' ,'AGE' ,'DIS' ,'RAD' ,'TAX' ,'PTRATIO',
 'B' ,'LSTAT']
data =pd.DataFrame(X_train, columns= columns)
data.insert(0, "MEDV", y_train)
data.head()
```

```python
sel_columns = ['LSTAT','INDUS','NOX','RM','MEDV']
data_sel = data[sel_columns]
import seaborn as sns
_ = sns.pairplot(data_sel, kind='reg', diag_kind='kde')
```

```python
import matplotlib.pyplot as plt

corr_coef = np.corrcoef(data_sel.values.T)
sns.set(font_scale=1.5)
heat_map = sns.heatmap(corr_coef,
                cbar=True,
                annot=True,
                square=True,
                fmt='.2f',
                annot_kws={'size':15},
                yticklabels=sel_columns,
                xticklabels=sel_columns)
plt.show()
```

We fit the training data to a linear model

```python
LinReg.fit(X_train, y_train)
```

We use the model to predict the test data

```python
y_predicted = LinReg.predict(X_test)
```

### Mean Square Error (MSE)
We measure the accuracy of the model by using the mean square error (MSE)

<img src=https://d1whtlypfis84e.cloudfront.net/guides/wp-content/uploads/2018/09/24042418/del7.jpg>
<img src="https://study.com/cimages/multimages/16/4e7cf150-0179-4d89-86f2-5cbb1f51c266_meansquarederrorformula.png">

Let us use sklearn metrics to get the Mean Square Error

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_true = y_test, y_pred = LinReg.predict(X_test)))
```

What is the error on the training data

```python
print(mean_squared_error(y_true = y_train, y_pred = LinReg.predict(X_train)))
```

```python
### How much does the model (the x variable) explain  the variability of the response data around its mean? This is called R^2
```

<img src="https://qph.fs.quoracdn.net/main-qimg-ef95453ec2e34729549271591ae1754c">

### R^2 for the model with all features

```python
LinReg.score(X_test, y_test)
```

### Mean Absolute Error (MAE)

<img src="https://miro.medium.com/max/630/1*OVlFLnMwHDx08PHzqlBDag.gif">


```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_true = y_test, y_pred = LinReg.predict(X_test)))
```

```python
print(LinReg.coef_)
```

Coefficients in multiple linear models represent the relationship between the given feature, X_i and the target, y, assuming that all the other features remain constant (conditional dependence). This is different from plotting  X_i versus y and fitting a linear relationship: in that case all possible values of the other features are taken into account in the estimation (marginal dependence).
See docu [scikit-learn](https://scikit-learn.org/stable/auto_examples/inspection/plot_linear_model_coefficient_interpretation.html#sphx-glr-auto-examples-inspection-plot-linear-model-coefficient-interpretation-py)

For example, y = 2*X_1 takes the other features into account; they are confounding the relationship. The slope may change if we control for the other features,e.g.,  y=3*X_1 + X_2 + 5*X_3. 

```python
print(LinReg.intercept_)
```

### Reducing overfitting and multi-collinearity

#### Ridge

```python
from sklearn.linear_model import Ridge
RidgeReg = Ridge(alpha=10)
RidgeReg.fit(X_train, y_train)
RidgeReg.score(X_test, y_test).  #R^2
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_true = y_test, y_pred = RidgeReg.predict(X_test)))
```

```python
print(mean_squared_error(y_true = y_train, y_pred = RidgeReg.predict(X_train)))
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_true = y_test, y_pred = RidgeReg.predict(X_test)))
```

#### Lasso

```python
from sklearn.linear_model import Lasso
LassoReg = Lasso(alpha=0.05)
LassoReg.fit(X_train, y_train)
LassoReg.score(X_test, y_test). # R^2
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_true = y_test, y_pred = LassoReg.predict(X_test)))
```

```python
print(mean_squared_error(y_true = y_train, y_pred = LassoReg.predict(X_train)))
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_true = y_test, y_pred = LassoReg.predict(X_test)))
```
